```python
import pandas as pd

# Charger votre fichier CSV
fichier = pd.read_csv("C:\\Users\\PC\\Downloads\\traitement_final.csv")

# Définir une fonction qui détermine la gravité
def determiner_gravite(row):
    if row['Nature_low'] == 1:
        return 'faible'
    elif row['Nature_medium'] == 1:
        return 'moyenne'
    else:
        # Si toutes les colonnes ont la valeur 0, on suppose que la gravité est élevée
        return 'élevée'

# Appliquer la fonction à chaque ligne du DataFrame pour créer la nouvelle colonne 'Gravité'
fichier['Gravité'] = fichier.apply(determiner_gravite, axis=1)

# Maintenant que la colonne 'Gravité' est créée, vous pouvez supprimer les anciennes colonnes si elles ne sont plus nécessaires
fichier = fichier.drop(['Nature_low', 'Nature_medium'], axis=1)

# Afficher les premières lignes pour vérifier le résultat
print(fichier.head())

# Continuer avec le traitement ou la sauvegarde du fichier modifié
fichier.to_csv("C:\\Users\\PC\\Downloads\\traitement_final_grav.csv", index=False)
```
